## 첫번째 설계
* `CircObj`, `RectObj`, `RTriObj` 클래스 작성 (3점)
* 같은 클래스의 인스턴스끼리 (`CircObj`끼리, `RectObj`끼리, `RTriObj`) 충돌 검사 테스트 사례 작성 및 각 테스트 사례를 svg 출력으로 확인 (3점)
* 서로 다른 클래스의 (`CircObj`와 `RectObj`, `RectObj`와 `RTriObj`, `RTriObj`와 `CircObj`) 인스턴스 충돌 검사 테스트 사례 작성 및 각 테스트 사례를 svg 출력으로 확인 (3점)

```java
abstract class ShapeObj { // 어떤 추상적인 모양의 물건 클래스
    Point p;   // 물건 위치에 대한 정보
    RGB color; // 물건의 색깔
    ShapeObj(Point p, RGB c) { this.p=p; color=c; }
    abstract Point closest(Point p); // 점 p와 가장 까까운 도형에 포함된 점 (소수점 이하는 적절히 처리해 정수 좌표로)
    abstract boolean collides(Point p); // 점 p와 충돌하는가?
    abstract boolean collides(ShapeObj o); // o와 충돌하는가?
    abstract String toSVG(); // svg 태그 생성
    @Override
    public String toString() { return toSVG(); }
}
```

```java
class CircObj extends ShapeObj { // 원 모양 물건 클래스
    // 원의 경우 p는 중심이다
    int r; // 반지름
    CircObj(int r, Point p, RGB c) {
        super(p, c); // 부모 클래스 생성자 호출
        this.r = r;
    }
    // 아래를 완성하라 (1점)
    @Override
    Point closest(Point p2) {
        double dist;
        double m = 0.0;
        double dtx, dty;
        long tx, ty;
        
        m = ( p.y - p2.y ) / ( p.x - p2.y ); //기울기 계산
        dist = Math.sqrt(Math.pow((p.x - p2.x),2) + Math.pow((p.y - p2.y),2) ); // 두 점간 거리 계산
        
        dtx = ( r / (1 + Math.pow(m,2)) ) + p.x;
        dty = ( r / (1 + Math.pow(m,2)) ) + p.y;
        
        tx = Math.round(dtx);
        ty = Math.round(dty);
        
        
        if(dist < r){
            return new Point( p2.x, p2.y );
        }else{
            return new Point( (int)tx, (int)ty );
        }
        
    }
    
    @Override
    boolean collides(Point p2) {
        
        double dist;
        
        dist = Math.sqrt(Math.pow((p.x - p2.x),2) + Math.pow((p.y - p2.y),2) );// 두 점간 거리 계산 
      
        if(dist <= r){
            return true;
        }else{
            return false;
        }
      
    }
    
    @Override
    boolean collides(ShapeObj o) {
        
        int tx, ty;
        
        tx = this.closest(o.p).x;
        ty = this.closest(o.p).y;
        
        Point tp = new Point(tx, ty);
        
        return o.collides(tp);
    }
    
    @Override
    String toSVG() {
        return "<circle cx='"+p.x+"' cy='"+p.y+"' r='"+r+"' style='stroke:black; stroke-width:1; fill:"+color+"; opacity:0.33' />";
    }
}
```

```java
public class RectObj extends ShapeObj { // 직사각형 모양 물건 클래스
    // 직사각형의 경우 경우 p는 왼쪽 위 꼭지점이다
    int w; // 너비 (양수)
    int h; // 높이 (양수)
    
    RectObj(int w, int h, Point p, RGB c) {
        super(p, c); // 부모 클래스 생성자 호출
        this.w = w;
        this.h = h;
    }

    @Override
    Point closest(Point p2) {
        
        int tx, ty;
        
        tx = p2.x;
        ty = p2.y;
        
        if(this.collides(p2)){
            return p2;
        }else{
            
            if(p2.x < p.x){
                tx = p.x;
            }else if(p2.x > p.x + w){
                tx = p.x + w;
            }
            
            if(p2.y < p.y ){
                ty = p.y;
            }else if(p2.y > p.y + h){
                ty = p.y + h;
            }
            
            return new Point(tx, ty);
            
        }
    }
        
    @Override
    boolean collides(Point p2) {
        
        if( p.x <= p2.x && p2.x <= p.x + w && p.y <= p2.y && p2.y <= p.y + h ){
            return true;
        }else{
            return false;
        }
        
    }
    
    @Override
    boolean collides(ShapeObj o) {
         int tx, ty;
        
        tx = this.closest(o.p).x;
        ty = this.closest(o.p).y;
        
        Point tp = new Point(tx, ty);
        
        return o.collides(tp);
    }
    
    @Override
    String toSVG() {
        return "<rect x='"+p.x+"' y='"+p.y+"' width='"+w+"' height='"+h+"' style='stroke:black; stroke-width:1; fill:"+color+"; opacity:0.33' />";
    }
}
```

```java
class RTriObj extends ShapeObj { // 직각삼각형 모양 물건 클래스
    int w; // 직각인 꼭지점으로부터 음수/양수로 왼쪽/오른쪽 방향으로의 w만큼 떨어진 꼭지점 나타냄
    int h; // 직각인 꼭지점으로부터 음수/양수로 아래쪽/위쪽 방향으로의 h만큼 떨어진 꼭지점 나타냄
    
    RTriObj(int w, int h, Point p, RGB c) {
        super(p, c); // 부모 클래스 생성자 호출
        this.w = w;
        this.h = h;
    }

    @Override
    Point closest(Point p2) {// 가까운 점 구하는 함수
        
        double m1, m2;
        double b1, b2;
        
        int ay = p.y + h;
        int bx = p.x + w;
        
        int tx = p2.x;
        int ty = p2.y;
        
        m1 = (ay - p.y) / (p.x - bx); // 빗변의 기울기
        m2 = -(1 / m1); //빗변의 법선 기울기
        
        b1 = p.y - ( m1 * bx ); // 빗변을 포함한 직선 방정식의 상수
        b2 = p2.y - ( m2 * p2.x ); // 법선의 직선 방정식의 상수
        
        double _x = (b2 - b1)/(m1 - m2);
        double _y = m1*((b2 -b1)/(m1 -m2)) + b1;
        
        long x = Math.round(_x);
        long y = Math.round(_y);
        
        if(collides(p2)){
            return p2;
        }else if(p2.x == p.x && p2.y > p.y + h ){
            ty = p.y + h;
        }else if(p2.y == p.y && p2.x > p.x + w ){
            tx = p.x + w;
        }else if( p.x > p2.x || p.y > p2.y){
            return p;
        }else{
            
            tx = (int)x;
            ty = (int)y;
        }
        
        return new Point(tx, ty); // 함수를 작성하라
    }
    @Override
    boolean collides(Point p2) {
        
        double dm;
        int y, b;
        long m;
        
        dm = ( ( p.y + h ) - p.y ) / (p.x - (p.x + w));
        
        m = Math.round(dm);
        b = p.y - (int)m * ( p.x + w );
        
        y = (int)m * p2.x + b;
           
        
         if( p.x <= p2.x && p2.x <= p.x + w && p.y <= p2.y && p2.y <= p.y + h ){
             
             if( y < p2.y){
                 return false;
             }else{
                 return true;
             }
             
        }else{
            return false;
        }
        
    }
    @Override
    boolean collides(ShapeObj o) {
        int tx, ty;
        
        tx = this.closest(o.p).x;
        ty = this.closest(o.p).y;
        
        Point tp = new Point(tx, ty);
        
        return o.collides(tp);
    }
    @Override
    String toSVG() {
        int pa = p.x + w;
        int pb = p.y + h;
        
        String points = "'" + p.x + ","+ p.y + " " + pa + "," + p.y + " " + p.x + "," + pb  + "'";
        
        return "<polygon points="+ points +"style='stroke:black; stroke-width:1; fill:"+color+"; opacity:0.33'/>";
    }

}
```

## 두번째 설계
첫번째 설계와 같은 문제를 해결하지만 조금 다른 방식의 설계로 프로그램을 재작성해 보라.
* `Circle`, `Rectangle`, `RTriangle` 클래스를 작성하라 (1점)
* 첫번째에서 했던 같은 테스트 사례들로 모두 실행해 보라 (3점)

```java
abstract class Shape { // 어떤 추상적인 모양 클래스
    RGB color; // 물건의 색깔
    Shape(RGB c) { color=c; }
    abstract String toSVG(Point self);
    abstract Point closest(Point self, Point p);
    abstract boolean collides(Point self, Point p);
    abstract boolean collides(Point self, Obj o);
}

class Obj {
    Point p;
    Shape s; // Shape이 Obj의 속성이다
    Obj(Point _p, Shape _s) { p=_p; s=_s; }
    Point closest(Point p) { return s.closest(this.p, p); } // 점 p와 가장 까까운 도형에 포함된 점 (소수점 이하는 적절히 처리해 정수 좌표로)
    boolean collides(Point p) { return s.collides(this.p, p); } // 점 p와 충돌하는가?
    boolean collides(Obj o) { return s.collides(this.p, o); } // o와 충돌하는가
    @Override
    public String toString() { return s.toSVG(this.p); }
}
```

```java
class Circle extends Shape { // 원 모양 클래스
    int r;
    Circle(int r, RGB c) { super(c); this.r = r; }
    @Override
    Point closest(Point self, Point p) {
        double dist;
        double m = 0.0;
        double dtx, dty;
        long tx, ty;
        
        m = ( self.y - p.y ) / ( self.x - p.x ); //기울기 계산
        dist = Math.sqrt(Math.pow((self.x - p.x),2) + Math.pow((self.y - p.y),2) ); // 두 점간 거리 계산
        
        dtx = ( r / (1 + Math.pow(m,2)) ) + self.x;
        dty = ( r / (1 + Math.pow(m,2)) ) + self.y;
        
        tx = Math.round(dtx);
        ty = Math.round(dty);
        
        
        if(dist < r){
            return new Point( p.x, p.y );
        }else{
            return new Point( (int)tx, (int)ty );
        }
    }
    @Override
    boolean collides(Point self, Point p) {
        double dist;
        
        dist = Math.sqrt(Math.pow((self.x - p.x),2) + Math.pow((self.y - p.y),2) );// 두 점간 거리 계산 
      
        if(dist <= r){
            return true;
        }else{
            return false;
        }
    }
    @Override
    boolean collides(Point self, Obj o) {
        int tx, ty;
        
        tx = this.closest(self, o.p).x;
        ty = this.closest(self, o.p).y;
        
        Point tp = new Point(tx, ty);
        
        return o.collides(tp);
    }
    @Override
    String toSVG(Point self) {
        return "<circle cx='"+self.x+"' cy='"+self.y+"' r='"+r+"' style='stroke:black; stroke-width:1; fill:"+color+"; opacity:0.33' />";
    }
}
```
